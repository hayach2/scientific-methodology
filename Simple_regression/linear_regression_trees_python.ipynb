```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.anova import anova_lm

def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals)
```

```python
data = pd.read_csv('arbres-tot.txt', sep=';')
data = data[data['HAUTEUR (m)'] != 0]
data
```

```python
circ = data['CIRCONFERENCE (cm)']
```

```python
height = data['HAUTEUR (m)']
```

```python
plt.scatter(circ, height, s=5)
plt.xlabel('circ')
plt.ylabel('height')
plt.show()
```

```python
height = np.array(height).reshape(-1, 1)
```

```python
reg = LinearRegression().fit(height, circ)
```

```python
print('model intercept :', reg.intercept_)
print('model coefficients : ', reg.coef_)
```

```python
# abline(reg.coef_, reg.intercept_)
intercept = reg.intercept_
slope = reg.coef_

plt.scatter(circ, height, s=5)
plt.xlabel('circ')
plt.ylabel('height')
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = intercept + slope * x_vals 
plt.plot(x_vals, y_vals)
    

plt.show()
```
